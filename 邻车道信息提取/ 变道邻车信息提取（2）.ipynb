```python
def repeat(re_list):
    a=[]
    for i in re_list:
        if i not in a:
            a.append(i)
    return a
```

```python
data_num=int(len(f_new)/41)
ID=f_new.Vehicle_ID.values
Preced=f_new.Preceding.values
Follow=f_new.Following.values
Frames=f_new.Frame_ID.values
ID=repeat(re_list=ID)
Frames=np.reshape(Frames,[-1,41])
Frames40=[]
for i in range(data_num):
    Frames40.append(Frames[i][:40])  
def pro_list(f,label,data_num):
    a=[]
    for i in range(data_num):
        a.append(f[label][i*41])
    return a
ID_list=pro_list(f_new,label='Vehicle_ID',data_num=data_num)
Preced_list=pro_list(f_new,label='Preceding',data_num=data_num)
Follow_list=pro_list(f_new,label='Following',data_num=data_num)
Frames40
```

```python
def choose_aim(PRE,FOL,FPS):
#     a=f[f.Vehicle_ID.isin([ID])]
    b=f[f.Vehicle_ID.isin([PRE])]
    b=b[b.Frame_ID.isin(FPS)]
    c=f[f.Vehicle_ID.isin([FOL])]
    c=c[c.Frame_ID.isin(FPS)]
    return b,c
```

```python
Pre = pd.DataFrame(columns=['Vehicle_ID','Frame_ID','Total_Frames','Global_Time','Local_X','Local_Y','Global_X','Global_Y','v_Length','v_Width','v_Class','v_Vel','v_Acc','Lane_ID','Preceding','Following','Space_Headway','Time_Headway'])
Fol = pd.DataFrame(columns=['Vehicle_ID','Frame_ID','Total_Frames','Global_Time','Local_X','Local_Y','Global_X','Global_Y','v_Length','v_Width','v_Class','v_Vel','v_Acc','Lane_ID','Preceding','Following','Space_Headway','Time_Headway'])
Pre1 = pd.DataFrame(columns=['Vehicle_ID','Frame_ID','Total_Frames','Global_Time','Local_X','Local_Y','Global_X','Global_Y','v_Length','v_Width','v_Class','v_Vel','v_Acc','Lane_ID','Preceding','Following','Space_Headway','Time_Headway'])
Fol1 = pd.DataFrame(columns=['Vehicle_ID','Frame_ID','Total_Frames','Global_Time','Local_X','Local_Y','Global_X','Global_Y','v_Length','v_Width','v_Class','v_Vel','v_Acc','Lane_ID','Preceding','Following','Space_Headway','Time_Headway'])
for i in range(data_num):
    a,b=choose_aim(PRE=Preced_list[i],FOL=Follow_list[i],FPS=Frames[i])
    a1,b1=choose_aim(PRE=Preced_list[i],FOL=Follow_list[i],FPS=Frames40[i])
    Pre=pd.concat([Pre,a], ignore_index=True)
    Fol=pd.concat([Fol,b], ignore_index=True)
    Pre1=pd.concat([Pre1,a1], ignore_index=True)
    Fol1=pd.concat([Fol1,b1], ignore_index=True)
d=pro_list(Pre,label='Vehicle_ID',data_num=int(len(Pre)/41))
e=pro_list(Fol,label='Vehicle_ID',data_num=int(len(Fol)/41))
```
