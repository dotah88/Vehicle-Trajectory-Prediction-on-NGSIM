```python
import numpy as np
import pandas as pd
import random
f=pd.read_csv("/home/mt/learn/NGSIMæ•°æ®å¤„ç†/csv/101/8:20-8:35/CL_train41.csv")
ff=pd.read_csv("/media/mt/Seagate Backup Plus Drive/csv/101/vehicle-trajectory-data/0820am-0835am/trajectories-0820am-0835am.csv")
num1=int(len(f)/41)
f=f.rename(columns={'Preceeding':'Preceding'})
f
```

```python
def choose_frame(arr):
    Pre = pd.DataFrame(columns=['Vehicle_ID','Frame_ID','Total_Frames','Global_Time','Local_X','Local_Y','Global_X','Global_Y','v_Length','v_Width','v_Class','v_Vel','v_Acc','Lane_ID','Preceding','Following','Space_Headway','Time_Headway','Vehicle_ID1'])
    Fol = pd.DataFrame(columns=['Vehicle_ID','Frame_ID','Total_Frames','Global_Time','Local_X','Local_Y','Global_X','Global_Y','v_Length','v_Width','v_Class','v_Vel','v_Acc','Lane_ID','Preceding','Following','Space_Headway','Time_Headway','Vehicle_ID1'])
    shape=np.shape(arr)
    for i in range(shape[0]):
        a=ff[ff.Vehicle_ID.isin([arr[i][2]])]
        b=a[a.Frame_ID.isin([arr[i][1]])]
        b['Vehicle_ID1']=arr[i][0]
        c=ff[ff.Vehicle_ID.isin([arr[i][3]])]
        d=c[c.Frame_ID.isin([arr[i][1]])]
        d['Vehicle_ID1']=arr[i][0]
        Pre=pd.concat([Pre,b], ignore_index=True)
        Fol=pd.concat([Fol,d], ignore_index=True)
    return Pre,Fol
    
```

```python
Pre1,Fol1=choose_frame(f4)
```

```python
pd1=pd.merge(f,Pre1, how='left', on=['Frame_ID','Preceding','Vehicle_ID'])
pd2=pd.merge(pd1,Fol1, how='left', on=['Frame_ID','Following','Vehicle_ID'])
```
