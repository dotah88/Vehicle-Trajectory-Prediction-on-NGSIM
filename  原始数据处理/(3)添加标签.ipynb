```python
CL_num=int(len(f2)/41)
obj_list=[41*i-1 for i in range(CL_num+1)]
obj_list=obj_list[1:]
obj_list
```

```python
TR_row=[]
for i in obj_list:
    if f2['Vehicle_ID'][i]==f2['Vehicle_ID'][i-1] and f2['Lane_ID'][i]==f2['Lane_ID'][i-1]+1:
        TR_row.extend(i-40+a for a in range(40))
TL_row=[]
for i in obj_list:
    if f2['Vehicle_ID'][i]==f2['Vehicle_ID'][i-1] and f2['Lane_ID'][i]==f2['Lane_ID'][i-1]-1:
        TL_row.extend(i-40+a for a in range(40))
TL_row  
```

```python
TR40=f2.loc[TR_row]
TR40['label']=2
TL40=f2.loc[TL_row]
TL40['label']=1
TR40
```

```python
df_combine = pd.concat([TR40, TL40], ignore_index=True)
df_combine.to_csv("/home/mt/learn/NGSIMæ•°æ®å¤„ç†/csv/101/8:20-8:35/RL_label.csv",index=False)
# TR40.to_csv("/home/mt/learn/NGSIMæ•°æ®å¤„ç†/csv/80/5:15-5:30/TR_label.csv",index=False)
```
